```python
# Import dependencies

import matplotlib.pyplot as plt
```

```python
# set the x-axis to a list of string for each month.

x_axis = ["Jan", "Feb", "Mar" ,"April", "May", "June", "July", "Aug", "Sept", "OCt", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month

y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot

plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() - object oriented interface
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend

plt.plot(x_axis, y_axis, marker="*", color="Blue", linewidth=2, label="Boston")

#create labels for the x and y axis

plt.xlabel("Date")
plt.ylabel("Fare($)")


# Set the y limit between 0 and 45

plt.ylim(0, 45)

# Create a title

plt.title("PyBer Fare by Month")

# Add a grid

plt.grid()

# Add the legend

plt.legend()
```

```python
# Create a plot using the Object Oriented approach

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis, color="Green", marker="*", linewidth=2, label = 'Boston')

# Add Title, axis labels, and ylimit

ax.set_title("PyBer Fare by  Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
ax.set_ylim(0, 45)

# Add grid lines

ax.grid()
```
