```python
# Import dependencies

import matplotlib.pyplot as plt
```

```python
# set the x-axis to a list of string for each month.

x_axis = ["Jan", "Feb", "Mar" ,"April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month

y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

# Line Graphs:

```python
# Create the plot

plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() - object oriented interface
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend

plt.plot(x_axis, y_axis, marker="*", color="Blue", linewidth=2, label="Boston")

#create labels for the x and y axis

plt.xlabel("Date")
plt.ylabel("Fare($)")


# Set the y limit between 0 and 45

plt.ylim(0, 45)

# Create a title

plt.title("PyBer Fare by Month")

# Add a grid

plt.grid()

# Add the legend

plt.legend()
```

```python
# Create a plot using the Object Oriented approach

fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis, color="Green", marker="*", linewidth=2, label = 'Boston')

ax.legend()

# Add Title, axis labels, and ylimit

ax.set_title("PyBer Fare by  Month")
ax.set_xlabel("Date")
ax.set_ylabel("Fare ($)")
ax.set_ylim(0, 45)

# Add grid lines

ax.grid()
```

# Bar Graphs:

```python
# Create a Bar graph

plt.bar(x_axis, y_axis, color="green", label="Boston")

# Create labels for the x and y axes

plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Create a title.

plt.title("PyBer Fare by Month")

# Add the legend.

plt.legend()
```

```python
# Create a horizontal bar graph

plt.barh(x_axis, y_axis, color="Magenta", label="Boston")

plt.gca().invert_yaxis()

plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.legend()
```

```python
# Create a horizontal bar graph using the object oriented interface method

fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="Cyan", label="Chicago")
ax.legend()

# Add title and axis labels
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")

# Invert the y-axis labels
ax.invert_yaxis()
```

# Scatter Plots:

```python
# Create a scatter plot using plt.plot()

plt.plot(x_axis, y_axis, 'o')
```

```python
# Create a scatter plot using plt.scatter()

plt.scatter(y_axis,x_axis,  c="Red", label="Chicago")
plt.legend()

# Add title and axis labels

plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Date")

# invert the y-axis

plt.gca().invert_yaxis()

# limit x-axis to certain range

plt.xlim(0,50)
```

```python
# new scatter plot with more dimensions (bubble plot)

plt.scatter(x_axis, y_axis, s=y_axis)

# Increase the size of the y-axis data:

y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

# Plot the larger y-axis data
plt.scatter(x_axis, y_axis, c="Blue", s=y_axis_larger)
```

```python
# Use list comprehesion to acheive the same as above with out the for loop

plt.scatter(x_axis, y_axis, s = [i*3 for i in y_axis])
```

```python
# Create a scatter plot using the onject oriented interface method

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a bubble chart using the object oriented interface method

fig, ax = plt.subplots()
ax.scatter( y_axis, x_axis, c="skyblue", s= [ i*5 for i in y_axis], 
           alpha = 0.8, edgecolor = "Black", linewidth = 2, label = "Boston")

# Add a legend
ax.legend()

# Add a title and axis labels
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")

# Limit the x-axis to a specific range
ax.set_xlim(0,50)

# invert the y-axis
ax.invert_yaxis()
```

# Pie Charts:

```python
# Create a pie chart

#plt.pie(y_axis, labels = x_axis)

# Explode the july data - create an explode list first; also format #'s to show as %

explode_values = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)

# add a color variable to specify the colors for each wedge

colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

# Increase the size of the pie so it is less crowded

plt.subplots(figsize=(8,8))

plt.pie(y_axis, labels = x_axis, 
        explode = explode_values, 
        autopct = '%.1f%%',
        colors = colors)

# plt.gca().invert_yaxis()

plt.show()
```

```python
# Create a pie chart using the object oriented interface method

fig, ax = plt.subplots(figsize=(8,8))

ax.pie(y_axis, labels=x_axis, 
       autopct = '%.1f%%',
       colors = colors,
    explode=explode_values,
      startangle=90,
      shadow=True,
      counterclock = False)

# ax.invert_yaxis()   --- use this or the counterclock argument

plt.show()
```

# Chart Extras - Error lines

### Line Graphs:

```python
%matplotlib inline
```

```python
# Import Dependancies

import matplotlib.pyplot as plt

import statistics
```

```python
# Set the x-axis to a list of strings for each month.

x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.

y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y_axis.

stdev = statistics.stdev(y_axis)

stdev
```

```python
# Plot the line graph with the stdev error bars using plt.errorbar and yerr= argument

plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)

```

```python
# Create the line graph with error bars using the object oriented approach

fig, ax = plt.subplots()

ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)

plt.show()
```

### Bar Graphs:

```python
# Create a bar graph with error line using the yerr= argument

plt.bar(x_axis,y_axis, yerr=stdev, capsize=3)
```

```python
# Create a bar graph w error lines using the object oriented approach

fig, ax = plt.subplots()

ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)

plt.show()
```

Adjust the tick spacing / marks

```python
# import numpy to make a range for the tick marks - use plt.xticks to specify range

import numpy as np

plt.barh(x_axis, y_axis)

plt.xticks(np.arange(0, 51, step=5.0))

plt.gca().invert_yaxis()
```

```python
# Create the same graph as above using the object oriented approach

fig, ax = plt.subplots()

ax.barh(x_axis, y_axis)

ax.set_xticks(np.arange(0, 51, step=5.0))

plt.show()
```

Add Minor Ticks:

```python
# Add minor ticks to a horizontal bar - import multiplelocator from matplotlib.ticker

from matplotlib.ticker import MultipleLocator

#Increase the size fi the plot figure.

fig, ax = plt.subplots(figsize=(8, 8))

ax.barh(x_axis, y_axis)

ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an incriment of 1.

ax.xaxis.set_minor_locator(MultipleLocator(1))

plt.show()
```
